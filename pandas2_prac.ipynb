```
data = {
    "Restaurant": ["R1", "R2", "R3", "R2", "R4", "R1"],
    "Votes": [120, 240, 150, 240, 80, 120],
    "Rating": [4.1, 4.5, 3.9, 4.5, 4.0, 4.1]
}

df = pd.DataFrame(data,index=["a","b","c","d","e","f"])
df
```

```
df.drop_duplicates() #drops the duplicate data
```

```
df.drop_duplicates(subset="Restaurant")
```

SQL Joins


```
import pandas as pd
# Sample DataFrames
samples_orders = pd.DataFrame({
    'product_id': ['P001', 'P002', 'P003'],
    'user_id': ['U001', 'U002', 'U003'],
    'review_id': ['R001', 'R002', 'R003']
})

samples_orders
```

```
samples_products = pd.DataFrame({
    'product_id': ['P001', 'P002', 'P004'],
    'product_name': ['Product A', 'Product B', 'Product D'],
    'category': ['Category 1', 'Category 2', 'Category 3']
})
samples_products
```

```
#Inner
pd.merge(samples_orders,samples_products,on="product_id",how="inner")
```

```
#Left Outer
pd.merge(samples_orders,samples_products,on="product_id",how="left")
```

```
#Right Outer
pd.merge(samples_orders,samples_products,on="product_id",how="right")
```

```
#Outer
pd.merge(samples_orders,samples_products,on="product_id",how="outer")
```

```
# Sample DataFrames
samples_orders = pd.DataFrame({
    'product_id': ['P001', 'P002', 'P003'],
    'user_id': ['U001', 'U002', 'U003'],
    'review_id': ['R001', 'R002', 'R003']
})

samples_products = pd.DataFrame({
    'id': ['P001', 'P002', 'P004'],
    'product_name': ['Product A', 'Product B', 'Product D'],
    'category': ['Category 1', 'Category 2', 'Category 3']
})
```

```
pd.merge(samples_orders,samples_products,left_on="product_id",right_on="id",how="inner")
```
